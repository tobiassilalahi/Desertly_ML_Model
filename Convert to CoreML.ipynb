```python
import os
import tensorflow as tf
import tfcoreml
from os import getcwd
from tensorflow.keras.models import load_model
```

```python
print(os.getcwd())
```

```python
keras_model = load_model('dessertly.h5')
input_name = keras_model.inputs[0].name.split(':')[0]
keras_output_node_name = keras_model.outputs[0].name.split(':')[0]
graph_output_node_name = keras_output_node_name.split('/')[-1]
output_labels = ['apple-pie', 'baklava', 'beignets', 'bread-pudding', 'carrot-cake', 
                 'cheesecake', 'chocolate-cake','chocolate-mousse', 'churros', 
                 'creme-brulee', 'cup-cakes', 'donuts', 'frozen-yogurt', 'hummus', 
                 'ice-cream', 'macarons', 'pancakes', 'panna-cotta', 'red-velvet-cake', 
                 'strawberry-shortcake', 'tiramisu', 'waffles']
model = tfcoreml.convert(tf_model_path='dessertly.h5',
                         image_input_names = ['input_2'],
                         input_name_shape_dict={'input_2': (1, 299, 299, 3)},
                         output_feature_names=[graph_output_node_name],
                         class_labels=output_labels, 
                         minimum_ios_deployment_target='13')
model.save('Dessertly.mlmodel')
```
